```python
from pandas.plotting import lag_plot
lag_plot(ahora_12.mendoza_indumentaria, lag=10)
```

Podemos observar que no existe una relación entre la cantidad de compras en el presente y la cantidad de compras hace 10 días atrás.
Al no depender entre sí, damos por entendido que al entrenar un modelo de predicción de una serie temporal para buscar cómo va a funcionar el ahora 12 en el futuro, me va a arrojar un error superior al 50%.

```python
#Entrenando el modelo para predecir cómo funcionará Ahora 12 en el futuro


train, test = ahora_12.mendoza_indumentaria[:-10], ahora_12.mendoza_indumentaria[-10:]

display(train)
display(test)
```

```python
from statsmodels.tsa.ar_model import AR
model = AR(train)
model_fit = model.fit(maxlag=26)
```

```python
pred = model_fit.predict(start=len(train), end=len(train)+len(test)-1, dynamic=False)

pd.DataFrame({
    'real': test,
    'prediction':pred,
    'diff': pred-test
})

error = (pred-test).abs().sum() / len(pred)

display(f"Este modelo tiene una cuota de error de {error} por lo que lo consideramos no aplicable.")
```
